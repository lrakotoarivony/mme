```python
import os
import shutil

# Path to the folder containing the images
train_folder = "/media/u1004/bae3ca51-3244-4f0e-9aa1-e2cd014c37af/dataset/imagenet/train/"

# Loop through each file in the train folder
for filename in os.listdir(train_folder):
    # Check if it's a file and matches the expected format
    if filename.endswith(".JPEG"):
        # Extract the directory prefix (e.g., n02119789)
        prefix = filename.split("_")[0]
        target_dir = os.path.join(train_folder, prefix)
        
        # Create the directory if it doesn't exist
        if not os.path.exists(target_dir):
            os.makedirs(target_dir)
        
        # Move the image to the corresponding directory
        source_path = os.path.join(train_folder, filename)
        target_path = os.path.join(target_dir, filename)
        shutil.move(source_path, target_path)

print("Images have been organized successfully!")
```

```python
import pickle

mapping_file = "datalists/mapping_imagenet.txt"
train_file = "datalists/imagenet2012_train_random_200k_reformatted.txt"

# Step 1: Parse mapping_imagenet.txt
line_to_reference = {}
with open(mapping_file, "r") as f:
    for i, line in enumerate(f, start=1):  # Line numbers start from 1
        reference, _ = line.strip().split(" ", 1)  # Split into reference and name
        line_to_reference[i - 1] = reference

# Step 2: Parse imagenet2012_train_random_200k_reformatted.txt
reference_to_class = {}
with open(train_file, "r") as f:
    for line in f:
        path, class_number = line.strip().split()  # Split into path and class_number
        # Extract reference from the file path
        reference = path.split("/")[1]  # Extract the reference part
        reference_to_class[reference] = class_number

# Step 3: Build the directory mapping line number to class
line_to_class = {}
for line_num, reference in line_to_reference.items():
    if reference in reference_to_class:
        line_to_class[line_num] = int(reference_to_class[reference])
    else:
        line_to_class[line_num] = None  # Handle cases where the reference isn't found

# Step 4: Print or save the directory
print(line_to_class)

with open('datalists/mapping.pkl', 'wb') as handle:
    pickle.dump(line_to_class, handle, protocol=pickle.HIGHEST_PROTOCOL)

# Optionally save the directory to a file
'''output_file = "datalists/line_to_class_mapping.txt"
with open(output_file, "w") as f:
    for line_num, class_number in line_to_class.items():
        f.write(f"{line_num} {class_number}\n")'''
```

```python
# Define the mapping of class names to class numbers
class_mapping = {
    "airplane": 0,
    "automobile": 1,
    "bird": 2,
    "cat": 3,
    "deer": 4,
    "dog": 5,
    "frog": 6,
    "horse": 7,
    "ship": 8,
    "truck": 9
}

# Open the input and output files
with open("datalists/cifar10_train.txt", "r") as infile, open("datalists/cifar10_train_reformatted.txt", "w") as outfile:
    for line in infile:
        # Remove any trailing whitespace
        line = line.strip()
        
        # Extract the class label from the path
        parts = line.split('/')
        class_name = parts[1]  # The second part should be the class name
        
        # Get the class number from the mapping
        class_number = class_mapping.get(class_name, -1)  # Default to -1 if not found
        
        # Write the new line to the output file
        outfile.write(f"{line} {class_number}\n")
```
